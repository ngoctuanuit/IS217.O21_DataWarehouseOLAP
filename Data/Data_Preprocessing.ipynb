```python
import pandas as pd
df = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='latin1')
df
```

```python
df['Customer Name'] = df['Customer Fname'] + ' ' + df['Customer Lname']
df['Customer Name'].head()
```

```python
df.columns
```

```python
columns = ['Order Id', 'Type', 'Order Status', 'order date (DateOrders)', 'Sales', 'Market', 'Order Item Discount Rate', 'Order Item Quantity',
           'Category Id', 'Category Name', 
           'Days for shipping (real)', 'Days for shipment (scheduled)', 'Delivery Status', 'Shipping Mode',
           'Customer Id', 'Customer Country', 'Customer Name', 'Customer Segment',
           'Product Card Id', 'Product Name', 'Product Price'
           ]
# Table: Orders, Customers, Categories, Prouducts, Shipping, Time
```

```python
df_res = df[columns]
df_res.head()
```

```python
df_res.info()
```

```python
df_res.dropna(inplace=True)
df_res.info()
```

```python
df_res.head()
```

```python
df_res.columns
```

```python
df_res = df_res.rename(columns={'Order Item Discount Rate':'discount_rate', 'Order Item Quantity': 'product_quantity', 'order date (DateOrders)': 'order_date', 'Days for shipping (real)': 'day_shipping_real', 'Days for shipment (scheduled)':'day_shipment_scheduled', 'Product Card Id': 'product_id'})
```

```python
df_res.columns
```

```python
df_res.columns = df_res.columns.str.lower().str.replace(' ', '_')
df_res
```
